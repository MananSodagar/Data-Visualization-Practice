```python
sns.set
sns.jointplot(x=tips["total_bill"], y=tips["tip"], kind="hex")
```

```python
sns.jointplot(x=tips["total_bill"], y=tips["tip"], kind="kde", shade=True)
```

```python
# f, ax= plt.subplot( figsize=(6,6))
cmap  =sns.cubehelix_palette(as_cmap=True, dark=0, light=1, reverse= True)
sns.kdeplot(x= "total_bill",y="tip", data= tips,cmap=cmap,nlevels=60, shade= True)
```

```python
g= sns.jointplot(kind='kde',x= "total_bill",y="tip", data= tips,color='m')
g.plot_joint(plt.scatter,color='y', linewidth=1, marker='+')
g.ax_joint.collections[0].set_alpha(1)
```
